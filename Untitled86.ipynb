```python
def nighbour(x, y):
    
    result = sorted(nx.common_neighbors(G,b[x],b[y]))
    
        
    return result
```

```python
def nighbour3(x, y,z):
    
    block1 = sorted(nx.common_neighbors(G,b[x],b[y]))
    block2=sorted(nx.common_neighbors(G,b[y],b[z]))
    import pandas as pd
    idx1 = pd.Index(block1)
    idx2 = pd.Index(block2)
    result=idx1.intersection(idx2)
    
        
    return result
```

```python
def nighbour4(x, y,z,w):
    
    block1 = sorted(nx.common_neighbors(G,b[x],b[y]))
    block2=sorted(nx.common_neighbors(G,b[w],b[z]))
    import pandas as pd
    idx1 = pd.Index(block1)
    idx2 = pd.Index(block2)
    result=idx1.intersection(idx2)
    
        
    return result
```

```python
def nighbour5(x, y,z,w,r):
    
    block1 = sorted(nx.common_neighbors(G,b[x],b[y]))
    block2=sorted(nx.common_neighbors(G,b[w],b[z]))
    block3=sorted(nx.common_neighbors(G,b[w],b[r]))
    import pandas as pd
    idx1 = pd.Index(block1)
    idx2 = pd.Index(block2)
    idx3=idx1.intersection(idx2)
    result=idx3.intersection(block3)
    
        
    return result
```
